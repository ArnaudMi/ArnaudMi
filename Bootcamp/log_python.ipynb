```python
def new_range(start, end, step) :
    while start <= end :
        yield start
        start += step

a = np.arange(1,20,1, dtype = np.int32)

for x in new_range(a[0], a[len(a)-1], 0.2) :
    if x > 12 and x < 14 :
        print(x)
```

```python
a = np.linspace(-2*np.pi,2*np.pi, 40)
b = np.sin(a)
c = np.cos(a)
new_x = np.linspace(0, 100000,100000)
d = np.log(new_x)

fig = plt.figure(figsize=(10,5), dpi = 120)
plt.subplot(2,1,1)
plt.grid()
plt.plot(a, b, label = 'sin', linewidth = 0.8, linestyle = '--')
plt.plot(a, c, label = 'cos', linewidth = 0.8, linestyle = '-')
plt.legend(loc = 'lower left')
plt.xticks([-np.pi*2,-np.pi,0,np.pi,2*np.pi],['$-2\pi$','$-\pi$','$0$','$\pi$','$2\pi$'])
ax1 = plt.gca()
ax1.spines['left'].set_color('none')
ax1.spines['right'].set_color('none')
ax1.spines['bottom'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.xaxis.set_ticks_position('bottom')
ax1.spines['bottom'].set_position(('data',0))
ax1.spines['bottom'].set_color('gray')
ax1.yaxis.set_ticks_position('left')
ax1.spines['left'].set_position(('data',0))
ax1.spines['left'].set_color('gray')




plt.subplot(2,2,3)
plt.plot(new_x, d, label = 'log')
plt.xscale('log')
plt.legend(loc = 'upper right')
ax2 = plt.gca()
ax2.spines['bottom'].set_color('none')
plt.subplot(2,2,4)
ax3 = plt.gca()
ax3.spines['top'].set_color('none')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pylab as plb

ax = Axes3D(plt.figure())
x = np.arange(-12, 6, 0.35)
y = np.arange(-12, 6, 0.35)
x, y = np.meshgrid(x, y)
k = np.sqrt(x**2 + y**2)
z = np.sin(k)

ax.plot_surface(x, y, z, rstride = 2, cstride = 1, cmap = plb.cm.gist_rainbow)
ax.contourf(x, y, z, zdir = 'z', offset = -3, cmap = plb.cm.gist_stern)
ax.set_zlim(-4,4)
plt.show()
```
